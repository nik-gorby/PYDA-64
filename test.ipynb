```python
def max_sq(min_size: int, max_size: int, steps=1):
    """
    Функция находит максимальный размер квадрата, на которые делится без 
    остатка прямоугольный участок с переданными размерами

    Args:
        min_size (int): меньший размер участка
        max_size (int): больший размер участка
        steps (int, optional): для подсчета шагов.

    Returns:
        int: Сторона наибольшего квадрата на которые делится участок без остатка
        int: Количество шагов в решении
    """

    if max_size % min_size == 0:
        return min_size, steps
    else:
        new_max_size = min_size
        new_min_size = max_size % min_size
        return max_sq(new_min_size, new_max_size, steps=steps + 1)


min_side_size = int(input('Введите меньшую сторону участка:'))
max_side_size = int(input('Введите большую сторону участка:'))

side_size, steps = max_sq(min_side_size, max_side_size)

print(f'Наибольшая сторона квадрата - {side_size}, шагов понадобилось - {
    steps}')
```

```python
import re
 
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
 
def counting(emails: list) -> dict:
    dic = {}
    for item in emails:
        match = re.search(r"@\w+\.\w+", item)
        if match:
            domain = match.group()[1:]
            dic.setdefault(domain, 0)
            dic[domain] += 1
    return dic
 
for k, v in counting(emails).items():
    print(k, v)
    
str(emails)
```
