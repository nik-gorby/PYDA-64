Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```

    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},

```

    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);

- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2
```

````python

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

command = ''

def search_by_number(doc_number):
    for item in documents:
        if item['number'] == doc_number:
            return True, item['type'], item['number'], item['name'] 
    return False, None

def shelf_search(doc_number):
    for k, v in directories.items():
        if doc_number in v:
            return k
    return None

def docs_info():
    for item in documents:
        shelf_number = shelf_search(item['number'])
        print(f"№: {item['number']}, тип: {item['type']}, владелец: {
              item['name']}, полка хранения: {shelf_number}")

def shelf_add():
    shelf_number = input('Введите номер полки:')
    if shelf_number not in directories.keys():
        directories[shelf_number] = []
        shelves = directories.keys()
        return f"Полка добавлена. Текущий перечень полок: {
            ', '.join(map(str, shelves))}."
    else:
        shelves = directories.keys()
        return f"Такая полка уже существует! Текущий перечень полок: {
            ', '.join(map(str, shelves))}."

def delete_shelf():
    shelf_number = input('Введите номер полки:')
    if shelf_number not in directories:
        print('Такой полки не существует! '
              f'Текущий перечень полок: {', '.join(map(str, 
              directories.keys()))}')
    elif directories[shelf_number]:
        print("На полке есть документы, удалите их перед удалением полки."
              f"Текущий перечень полок:{', '.join(map(str, 
              directories.keys()))}")
    else:
        del directories[shelf_number]
        print('Полка удалена. Текущий перечень полок: '
              f'{', '.join(map(str, directories.keys()))}')
        
def doc_add():
    doc_number = input('Введите номер документа:')# номер
    doc_type = input('Введите тип документа:')# тип
    doc_owner = input('Введите владельца документа:')# владелец
    doc_directory = input('Введите номер полки для хранения документа:')# полка
    
    if doc_directory in directories:
        # 'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'
        documents.append({'type': doc_type, 
                          'number': doc_number, 
                          'name': doc_owner})
        directories[doc_directory].append(doc_number)
        print('Документ добавлен.')
        docs_info()
    else:
        print('Такой полки не существует. Добавьте полку командой ads.')
        docs_info()
        
def doc_del(doc_number=None):
    if not doc_number: doc_number = input('Введите номер документа:')
    if search_by_number(doc_number)[0]:
        # delete doc
        for index, doc in enumerate(documents):
            if doc['number'] == doc_number:
                del documents[index]
        # delete doc from directory
        for directory in directories:
            if doc_number in directory:
                directory.remove(doc_number)
        return True                
    else: 
        return None
    
def doc_move(doc_number=None):
    if not doc_number: doc_number = input('Введите номер документа:')
    new_directory_number = input('Введите номер полки:')
    if new_directory_number not in directories:
        print('Такой полки не существует!'
              f"Текущий перечень полок:{', '.join(map(str, 
              directories.keys()))}")
        return None
    elif not search_by_number(doc_number)[0]:
        print('Документ не найден в базе\nТекущий перечень документов:')
        docs_info()
        return None
    else:
        for directory in directories:
            directories[directory] = [
                '' if item == doc_number else item for item in 
                directories[directory]
            ]
        directories[new_directory_number].append(doc_number)
        return True

while command != 'q':
    command = input('Введите команду:')
    if command == 'p':
        doc_number = input('Введите номер документа: ')
        search_result, doc_owner = search_by_number(doc_number)
        if search_result:
            print(f"Владелец документа: {doc_owner}")
        else: print('Документ не найден в базе')
    elif command == 's':
        doc_number = input('Введите номер документа: ')
        print(f'Документ хранится на полке: {shelf_search(doc_number)}')
    elif command == 'l':
        docs_info()
    elif command == 'ads':
        print(shelf_add())
    elif command == 'ds':
        delete_shelf()
    elif command == 'ad':
        doc_add()
    elif command == 'd':
        if doc_del():
            print('Документ удален.\nТекущий перечень документов:')
            docs_info()
        else: 
            print('Документ не найден в базе.\nТекущий перечень документов:')
            docs_info()
    elif command == 'm':
        result = doc_move()
        if result:
            print('Документ перемещен.\nТекущий перечень документов:')
            docs_info()
````
